### Test notebook for Downloading a globus public zip file and unzip it

```python
# replace this URL in quotes with the one you want to download, and run cell
file_url = ""
```

```python
from pathlib import Path
from urllib.request import urlretrieve
from urllib.parse import urlparse
import zipfile

# Test case: Check if file_url is empty
if not file_url or not file_url.strip():
    raise ValueError("file_url is empty! Please set a valid URL in Cell 2 before running this cell.")

# Uses the URL stored in file_url (defined in Cell 2)
data_dir = Path("data")
data_dir.mkdir(parents=True, exist_ok=True)

file_name = Path(urlparse(file_url).path).name or "download.zip"
zip_path = data_dir / file_name
urlretrieve(file_url, zip_path)

with zipfile.ZipFile(zip_path, "r") as zf:
    zf.extractall(data_dir)

print(f"Downloaded to: {zip_path}")
print(f"Extracted into: {data_dir} folder")
```
